```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
file_path = r'C:/Users\Gregory Bazuaye/Downloads/BankNote_Authentication.csv'

dataset = pd.read_csv(file_path)

dataset.head(10)
```

```python
x = dataset.iloc[: , : -1]
y = dataset.iloc [: , -1]


x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.25)
```

```python
classifier = RandomForestClassifier()

classifier.fit(x_train , y_train)

predictions = classifier.predict(x_test)


accuracy = accuracy_score(predictions , y_test)
conf_matrix = confusion_matrix(predictions , y_test)


columns = ['Actual True' , 'Actual False']
indices = ['Predicted True' , 'Predicted False ']

print(f'The acuracy is {accuracy}')

conf_matrix = pd.DataFrame(conf_matrix , columns = columns , index = indices)

conf_matrix

```

```python
import pickle
pickle_out = open('classifier.pkl' , 'wb')
pickle.dump(classifier , pickle_out)
pickle_out.close()

```
